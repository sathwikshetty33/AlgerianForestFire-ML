```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"] = 0
df.loc[123:,"Region"] = 1
```

```python
df.info()
```

```python
df["Region"]=df["Region"].astype(int)
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]= df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
obj =[fea for fea in df.columns if df[fea].dtype == 'O']
```

```python
for o in obj:
    if o!= "Classes":
        df[o] = df[o].astype(float)
```

```python
df.info()
```

```python
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index=False)
```

```python
d = df
```

```python
d.drop(columns=['day', 'month', 'year'], inplace=True)
d.head()
```

```python
d['classes']=np.where(d['Classes'].str.contains('not fire'), 0, 1)
```

```python
d.head()
```

```python
d.classes.value_counts()
```

```python
plt.style.use('seaborn-v0_8')
d.hist(bins=50, figsize=(20, 15), color='blue', alpha=0.5)
plt.show()
```

```python
per = d['classes'].value_counts(normalize=True)*100
```

```python
classlabels = ['not fire', 'fire']
plt.figure(figsize=(10, 5))
plt.pie(per,labels=classlabels,autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.show()
```
